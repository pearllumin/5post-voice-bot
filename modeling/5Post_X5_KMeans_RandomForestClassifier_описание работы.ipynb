# Описание работы ноутбука 5Post_X5_KMeans_RandomForestClassifier #

---

## 1 ЭТАП ##
### Изучение, обработка и очистка данных ###

Первый этап нужен для того чтобы изучить предоставленные данные, обработать и очистить их.
На первом этапе мы проделываем следующие операции:

1. Изучили структуру датасета
2. Узнали и изучили данные в столбцах
3. Приняли решение какие данные надо оставить
4. Удалили лишние столбцы и дубликаты
5. Отчистили текстовые данные от лишних знаков препинания и привели буквы к нижнему регистру
6. Изучаем тексты длиной меньше 2 слов. Их количество не велико, но в них есть важные тексты, которые в дальнейшем будут влиять на классификацию, поэтому не удаляем.
7. Принято решение проанализировать очень длинные тексты, т.к. в них могут быть признаки более 1 класса. В нашей модели не предусмотрена классификация одного текста по двум признакам. Для анализа и удаления длинных текстов мы использовали IQR - метод межквартильного размаха.
8. Длинные тексты сохранили в отдельный датасет и потом их используем с классом "Соедините с оператором"

---

## 2 ЭТАП ##
###  Обучение классификатора на основе K-Means ###

Во втором этапе приступаем к созданию модели классификатора на основе метода k-средних K-Means.
1. Преобразуем текст в матрицу TF-IDF признаков. Выбрали 120 000 признаков, это порядка 30% от чистого датасета.
2. Далее с помощью метода локтя пробуем найти излом дл определения количества кластеров, которое будем использовать в K-Means.
3. Инициализируем и обучаем K-Means модель, указав 39 кластеров.
4. Визуализируем кластеры в виде облаков слов и гситограмм с распределением слов по частоте использования.
5. Определяем ручным способом названия классов.
6. Если кластер не даёт точного названия класса, разбиваем кластер на подкластеры.
7. Обучаем и сохраняем нашу модель классификатора.
8. Создаём функцию для разметки нашего датасета.
9. Используя нашу функцию и обученную модель размечаем наш датасет.

---

## 3 ЭТАП. ##
### Изучение, обработка, очистка и слияние размеченного датасета ###

1. Обрабатываем датасет удаляя лишние данные.
2. Смотрим отсортированное количество предсказанных классов.
3. Классы с малым количеством перемещаем в класс Unknown Cluster.
4. Производим слияние классифицированного датасета с датасетом с длинными текстами по признаку "Соедините с оператором"
5. Сохраняем классифицированный датасет.

---

## 4 ЭТАП. ##
### Обучение модели классификатора RandomForestClassifier ###

1. Разделяем наши данные на train и test
2. Преобразовываем текст в числовые данные с помощью Tfidf
3. Обучаем модель
4. Оцениваем модель и получаем метрики оценки классификационной модели.
5. Сохраням моедль и тестируем.
